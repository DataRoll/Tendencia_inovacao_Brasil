```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

```python
#inovacao, predicao, tendencia, indicador, previsao, prognostico, innovation, forecast, prediction, trend, indicator, prognosis
inovacao=pd.read_csv('Dados palavra chave/inovação_scopus.csv', sep=';')
predicao=pd.read_csv('Dados palavra chave/predição_scopus.csv', sep=';')
tendencia=pd.read_csv('Dados palavra chave/tendência_scopus.csv', sep=';')
indicador=pd.read_csv('Dados palavra chave/indicador_scopus.csv', sep=';')
previsao=pd.read_csv('Dados palavra chave/previsão_scopus.csv', sep=';')
prognostico=pd.read_csv('Dados palavra chave/prognostico_scopus.csv', sep=';')
innovation=pd.read_csv('Dados palavra chave/innovation_scopus.csv', sep=';')
forecast=pd.read_csv('Dados palavra chave/forecast_scopus.csv', sep=';')
prediction=pd.read_csv('Dados palavra chave/prediction_scopus.csv', sep=';')
trend=pd.read_csv('Dados palavra chave/trend_scopus.csv', sep=';')
indicator=pd.read_csv('Dados palavra chave/indicator_scopus.csv', sep=';')
prognosis=pd.read_csv('Dados palavra chave/prognosis_scopus.csv', sep=';')
```

```python
plt.figure(figsize=(20,16))

plt.title('Numero de publicações por ano na revista Scopus', fontdict={'fontweight':'bold', 'fontsize': 15})

plt.plot(predicao.years, predicao.publications, 'g.-', label='predicao')
plt.plot(tendencia.years, tendencia.publications, 'r.-', label='tendencia')
plt.plot(inovacao.years, inovacao.publications, 'b.-', label='inovacao')
plt.plot(indicador.years, indicador.publications, 'g.-', label='indicador')
plt.plot(previsao.years, previsao.publications, 'b.-', label='previsao')
plt.plot(prognostico.years, prognostico.publications, 'm.-', label='prognostico')
plt.plot(innovation.years, innovation.publications, 'k.-', label='innovation')
plt.plot(forecast.years, forecast.publications, 'r.-', label='forecast')
plt.plot(prediction.years, prediction.publications, 'k.-', label='prediction')
plt.plot(trend.years, trend.publications, 'm.-', label='trend')
plt.plot(indicator.years, indicator.publications, 'b.-', label='indicator')
plt.plot(prognosis.years, prognosis.publications, 'y.-', label='prognosis')


plt.xlabel('Year')
plt.ylabel('publications')
plt.xscale("linear")
plt.yscale("linear")

plt.legend()

plt.show()
```
